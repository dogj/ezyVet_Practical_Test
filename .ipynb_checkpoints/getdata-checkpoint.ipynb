```python
import pandas as pd
import numpy as np
import json,requests
import xlwt
from openpyxl import load_workbook
import  glob
```

```python
# key word list
keyword_list=['Tobacco',"Alcohol","Gambling","Weapons","Adult Entertainment","Palm Oil","Fossil Fuels","Human Rights","Animal Welfare","GMOs"]


def validate(str1,str2,percentage):
    if str(str1)=="NAN":
        return False
    if percentage<0.0001:
        return False
    if str1 in exclude_list and str1!=str2:
        return False
    if str1 in include_list and str1==str2:
        return True  
    if len(str1)>5 and str1 in str2:
        return True    
    if len(str2)>8 and str2 in str1:
        return True
```

```python
#all excel


fund_id=[]
fund_name=[]
fund_offer=[]

#define score for the bad company list

Tobacco_score=[]
Alcohol_score=[]
Gambling_score=[]
Weapons_score=[]
Adult_score=[]
Palm_Oil_score=[]
Fossil_Fuels_score=[]
Human_Rights_score=[]
Animal_Welfare_score=[]
GMOs_score=[]
Summary_score=[]
Summary_fund=[]
finished=[]



#read all xls file in bad folder
for f in glob.glob('../bad/*.xlsx'):
    bad_list = pd.read_excel(f)
    df_bad_list=pd.DataFrame(bad_list)

df_bad_list= df_bad_list.apply(lambda x: x.astype(str).str.upper())


for f in glob.glob('../special companies/too short/*.xlsx'):
    special_list = pd.read_excel(f)
    df_special_list=pd.DataFrame(special_list)
df_special_list= df_special_list.apply(lambda x: x.astype(str).str.upper())

exclude_data = np.array(df_special_list['exclude'])#np.ndarray()
exclude_list=exclude_data.tolist()#list
exclude_list = [x for x in exclude_list if str(x) != 'NAN']

include_data = np.array(df_special_list['include'])#np.ndarray()
include_list=include_data.tolist()#list
include_list = [x for x in include_list if str(x) != 'NAN']


#read all xlsx in fund folder
for f in glob.glob('../fund/*.xlsx'):
    df2 = pd.read_excel(f)         # read excel
    df=pd.DataFrame(df2)
    df.columns=[1,2,3,4]
    
    #change column names read fund id, name and offer 
    fund_id.append(df.loc[1,4])
    fund_name.append(df.loc[1,2])
    fund_offer.append(df.loc[0,4])
    
    #drop unwanted columns and rows
    df=df.drop([0,1,2,3])
    df=df.drop(columns=[3, 4])
    
    #for other cases 
    if "%" in str(df.loc[4,2]):
        p_float = df[2].str.strip("%").astype(float)/100
        df[2]=p_float.round(decimals=6)
        print("changed")
    if df.loc[4,2]>1:
        print(fund_id[-1])
        print("this fund source data structure has a little problem but fixed")
        df[2] = df[2]/100
        
    
    current_Tobacco_score=0
    current_Alcohol_score=0
    current_Gambling_score=0
    current_Weapons_score=0
    current_Adult_score=0
    current_Palm_Oil_score=0
    current_Fossil_Fuels_score=0
    current_Human_Rights_score=0
    current_Animal_Welfare_score=0
    current_GMOs_score=0
    current_Tobacco_summary=''
    current_Alcohol_summary=''
    current_Gambling_summary=''
    current_Weapons_summary=''
    current_Adult_summary=''
    current_Palm_Oil_summary=''
    current_Fossil_Fuels_summary=''
    current_Human_Rights_summary=''
    current_Animal_Welfare_summary=''
    current_GMOs_summary=''

    current_summary=''
    
    for i,each in df[1].items():
        for bad_list in df_bad_list['Tobacco'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):        
                current_Tobacco_score= current_Tobacco_score + df[2][i]
                current_Tobacco_summary=current_Tobacco_summary+str(df[1][i])+'---'+format(df[2][i], '.2%')+  ' '
                break            
        for bad_list in df_bad_list['Alcohol'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):    
                current_Alcohol_score= current_Alcohol_score + df[2][i]
                current_Alcohol_summary=current_Alcohol_summary+str(df[1][i])+'---'+format(df[2][i], '.2%')+  ' '
                break        
        for bad_list in df_bad_list['Gambling'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):    
                current_Gambling_score= current_Gambling_score + df[2][i]
                current_Gambling_summary=current_Gambling_summary+str(df[1][i])+"---"+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Weapons'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):    
                current_Weapons_score= current_Weapons_score + df[2][i]
                current_Weapons_summary=current_Weapons_summary+str(df[1][i])+ '---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Adult Entertainment'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_Adult_score= current_Adult_score + df[2][i]
                current_Adult_summary=current_Adult_summary+str(df[1][i])+'---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Palm Oil'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_Palm_Oil_score= current_Palm_Oil_score + df[2][i]
                current_Palm_Oil_summary=current_Palm_Oil_summary+str(df[1][i])+ '---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Fossil Fuels'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_Fossil_Fuels_score= current_Fossil_Fuels_score + df[2][i]
                current_Fossil_Fuels_summary=current_Fossil_Fuels_summary+str(df[1][i])+'---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Human Rights'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_Human_Rights_score= current_Human_Rights_score + df[2][i]
                current_Human_Rights_summary=current_Human_Rights_summary+str(df[1][i])+ '---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['Animal Welfare'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_Animal_Welfare_score= current_Animal_Welfare_score + df[2][i]
                current_Animal_Welfare_summary=current_Animal_Welfare_summary+str(df[1][i])+ '---'+format(df[2][i], '.2%')+  ' '
                break
        for bad_list in df_bad_list['GMOs'].unique():
            str1=str(bad_list).upper()
            str2=str(each).upper()
            if validate(str1,str2,df[2][i]):
                current_GMOs_score= current_GMOs_score + df[2][i]
                current_GMOs_summary=current_GMOs_summary+str(df[1][i])+ '---'+format(df[2][i], '.2%')+  ' '
                break

    summary_list=[current_Tobacco_summary,current_Alcohol_summary,current_Gambling_summary,current_Weapons_summary,
                current_Adult_summary,current_Palm_Oil_summary,current_Fossil_Fuels_summary,current_Human_Rights_summary ,current_Animal_Welfare_summary
                 ,current_GMOs_summary]
    for i,each in  enumerate(summary_list):
        if each!='':
            summary_list[i]=keyword_list[i]+" : "+each+'\n'
                    
                    
    current_summary= "".join(summary_list)

    Tobacco_score.append(current_Tobacco_score)
    Alcohol_score.append(current_Alcohol_score)
    Gambling_score.append(current_Gambling_score)
    Weapons_score.append(current_Weapons_score)  
    Adult_score.append(current_Adult_score)
    Palm_Oil_score.append(current_Palm_Oil_score)
    Fossil_Fuels_score.append(current_Fossil_Fuels_score)
    Human_Rights_score.append(current_Human_Rights_score)
    Animal_Welfare_score.append(current_Animal_Welfare_score)
    GMOs_score.append(current_GMOs_score)
    Summary_fund.append(current_summary)
    
    
    file=str('../cleaned/'+fund_id[-1]+'__'+fund_name[-1].replace("/","_")+'__'+fund_offer[-1]+'.xlsx')
    writer = pd.ExcelWriter(file, engine='xlsxwriter')
    df.to_excel(writer, sheet_name='Sheet1')
    writer.save()

print("all_excel done")
print(len(fund_id))
```
